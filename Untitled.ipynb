```python
import pickle as pkl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import matplotlib 
matplotlib.rcParams.update({'font.size': 22})
```

```python
dipole_results = pd.read_pickle("dipole_mags/dipole_mags.pkl")
dipole_results
```

```python
previous_method = pd.read_pickle("gfn1/gfn1_eigdiff_test_no_RMSE.pkl")
previous_method
```

```python
fig, ax = plt.subplots(1, 2) 

l1 = ax[0].scatter(dipole_results["tddft_energy"].to_numpy(), dipole_results["xtb_energy"].to_numpy(), color='red', marker='x')
l2 = ax[0].scatter(previous_method["tddft_energy"].to_numpy(), previous_method["xtb_energy"].to_numpy(), color='black', marker='x')
l5, = ax[0].plot(dipole_results["tddft_energy"].to_numpy(), dipole_results["tddft_energy"].to_numpy(), linestyle='--', color='grey')

l3 = ax[1].scatter([np.linalg.norm(x) for x in dipole_results["tddft_dipoles"]], [np.linalg.norm(x) for x in dipole_results["xtb_dipoles"]], color='red', marker='x')
l4 = ax[1].scatter([np.linalg.norm(x) for x in previous_method["tddft_dipoles"]], [np.linalg.norm(x) for x in previous_method["xtb_dipoles"]], color='black', marker='x')
ax[1].plot([np.linalg.norm(x) for x in previous_method["tddft_dipoles"]], [np.linalg.norm(x) for x in previous_method["tddft_dipoles"]], linestyle='--', color='grey')

ax[0].set_xlabel("TD-DFT")
ax[0].set_ylabel("xTB methods")

ax[1].set_xlabel("TD-DFT")
ax[1].set_ylabel("xTB methods")

ax[0].set_title("Excitation energy / eV")
ax[1].set_title("Transition dipole magnitude / au")

fig.legend((l1, l2, l5), ("Density Matrix scaling", "No scaling", "y=x"), 'center right')

fig.set_size_inches(24,12)
```
